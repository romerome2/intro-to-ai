```python
import os
path = "train/"
training_data = os.listdir(path)
print(training_data)
```

```python
labels = []
for y in training_data:
    if 'cat' in y:
        labels.append(0)
    else:
            labels.append(1)
```

```python
import cv2
import numpy as np
cat_dog_images = []
for file_name in training_data:
    img = cv2.imread('resized/' + file_name)
    cat_dog_images.append(img)

cat_dog_images = np.asarray(cat_dog_images)
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as img

cat_iamge = img.imread("train/cat.100.jpg")
plt.imshow(cat_iamge)
```

```python
from sklearn.model_selection import train_test_split

labels = np.asarray(labels)

x_train, x_test, y_train, y_test = train_test_split(cat_dog_images, labels, test_size=0.2, random_state=2)

x_train_scaled = x_train / 225
x_test_scaled = x_test / 225
```

```python
import tensorflow as tf
import tensorflow_hub as hub


mobilenet_model = 'https://tfhub.dev/google/tf2-preview/mobilenet_v2/feature_vector/4'

pretrained_model = hub.KerasLayer(mobilenet_model, input_shape=(224,224,3), trainable=False)
```

```python
model = tf.keras.Sequential([
    pretrained_model,
    tf.keras.layers.Dense(2)
])

model.summary()
```

```python
model.compile(
    optimizer = 'adam',
    loss = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
    metrics = ['acc']
)
```

```python
model.fit(x_test_scaled, y_train, epochs= 5)
```
